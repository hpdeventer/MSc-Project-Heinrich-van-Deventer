```python
from models_and_procedures_definitions import *
import numpy as np
import matplotlib.pyplot as plt

# Define the target function
def f(x1, x2):
    return np.sin(4 * np.pi * x1) * np.sin(4 * np.pi * x2)

# Create a grid of points in the domain [0, 1]^2
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
X1, X2 = np.meshgrid(x1, x2)

# Compute the function values at the grid points
Z = f(X1, X2)

# Plot the target function using imshow
plt.imshow(Z, extent=[0, 1, 0, 1], origin='lower', cmap='viridis')
plt.colorbar(label='$f(x) = sin(4 \pi x_1) \cdot sin(4 \pi x_2)$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Plot of $f(x) = sin(4 \pi x_1) \cdot sin(4 \pi x_2)$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.losses import MeanAbsoluteError

models = initialize_all_models(input_dimension=2, seed_val=1)
compile_models(models)

# Generate the training data
X_train = np.random.uniform(0, 1, (10000, 2))
y_train = f(X_train[:, 0], X_train[:, 1])

# Define the loss function
loss_fn = MeanAbsoluteError()

# Initialize list to store history of losses for each model
histories = []

# Fit each model to the training data
for model, name in models:
    print(f"Training {name}...")
    history = model.fit(X_train, y_train, epochs=100, batch_size=100)
    histories.append(history)
    
```

```python
# Flatten the meshgrid arrays and stack them to create a 2D array of points
X_test = np.stack([X1.flatten(), X2.flatten()]).T

# Predict the model output on the grid of points
for model, name in models:
    print(f"Predicting with {name}...")
    y_pred = model.predict(X_test)

    # Reshape the prediction array to have the same shape as X1 and X2
    Z_pred = y_pred.reshape(X1.shape)

    # Plot the predicted function using imshow
    plt.imshow(Z_pred, extent=[0, 1, 0, 1], origin='lower', cmap='viridis')
    plt.colorbar(label=f'$f(x) = {name}(x)$')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.title(f'Plot of $f(x) = {name}(x)$')
    plt.show()
```

```python
# Predict the model output on the grid of points
fig, axs = plt.subplots(len(models), 1, figsize=(10, len(models)*5), sharex=True, sharey=True)

for (model, name), ax in zip(models, axs.flatten()):
    print(f"Predicting with {name}...")
    y_pred = model.predict(X_test)

    # Reshape the prediction array to have the same shape as X1 and X2
    Z_pred = y_pred.reshape(X1.shape)

    # Plot the predicted function using imshow
    im = ax.imshow(Z_pred, extent=[0, 1, 0, 1], origin='lower', cmap='viridis')
    
    # Add a colorbar to each subplot
    fig.colorbar(im, ax=ax, label=f'$f(x) = {name}(x)$')
    
    ax.set_title(f'Plot of $f(x) = {name}(x)$')

axs[-1].set_xlabel('$x_1$')  # only set x-label for the bottom subplot
for ax in axs:
    ax.set_ylabel('$x_2$')

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Predict the model output on the grid of points
fig, axs = plt.subplots(1, len(models), figsize=(len(models)*3, 5), sharex=True, sharey=True)

# Initialize variables to store the minimum and maximum of all predictions
vmin = np.inf
vmax = -np.inf

# List to store all predictions
all_pred = []

for (model, name), ax in zip(models, axs.flatten()):
    print(f"Predicting with {name}...")
    y_pred = model.predict(X_test)

    # Update vmin and vmax
    vmin = min(vmin, y_pred.min())
    vmax = max(vmax, y_pred.max())

    # Reshape the prediction array to have the same shape as X1 and X2
    Z_pred = y_pred.reshape(X1.shape)
    
    # Store Z_pred in list
    all_pred.append(Z_pred)

for (model, name), ax, Z_pred in zip(models, axs.flatten(), all_pred):
    
    # Plot the predicted function using imshow with shared color limits
    im = ax.imshow(Z_pred, extent=[0, 1, 0, 1], origin='lower', cmap='viridis', vmin=vmin, vmax=vmax)

    ax.set_title(name)

for ax in axs:
    ax.set_xlabel('$x_1$') 

axs[0].set_ylabel('$x_2$')  # only set y-label for the first subplot

# Create a divider for the existing axes instance.
divider = make_axes_locatable(ax)
    
# Add an axes to the right of the main axes.
cax = divider.append_axes("right", size="5%", pad=0.05)

# Add a colorbar to the last subplot with adjusted height
fig.colorbar(im, cax=cax)

plt.tight_layout()
plt.show()
```
